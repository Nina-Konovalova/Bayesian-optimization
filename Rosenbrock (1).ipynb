```
import numpy as np
import scipy 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def Rosenbrock (x):
      return (1-x[0])**2 + 100.0*(x[1] - x[0]**2)**2

#x = np.arange(-3,3.1,0.1)
x = np.arange(-3, 3, 0.15)
grid_x, grid_y = np.meshgrid(x,x)

surf = np.zeros((grid_x.shape[1], grid_x.shape[1]))
for i in range (grid_x.shape[1]):
  for j in range (grid_x.shape[1]):
     surf[i][j] = Rosenbrock([grid_x[i][j],grid_y[i][j]])

from matplotlib import cm
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(grid_x, grid_y, surf, cmap=plt.cm.Spectral, linewidth=0, antialiased=False)
plt.show()

np.random.seed(145)
x_train = np.random.uniform(-3, 3, (50, 2))
plt.scatter(x_train[:,0],x_train[:,1], color = 'green')
plt.show()
y_train = np.zeros((x_train.shape[0]), dtype = float) #!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
for i in range (len(x_train[:,0])):
   y_train[i] = Rosenbrock([x_train[i][0], x_train[i][1]])
%matplotlib inline
from matplotlib import cm
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.scatter(x_train[:,0],x_train[:,1],y_train, color = 'red', s=40, marker = '+')
ax.plot_surface(grid_x, grid_y, surf, cmap=cm.viridis, linewidth=0, antialiased=False, alpha = 0.2)

fig.savefig('my_figure.png')
plt.show()

from scipy.optimize import minimize
#np.random.seed(799351)
#x0 = np.array([np.random.choice(np.arange(-2.9,2.9,0.1)),np.random.choice(np.arange(-2.9,2.9,0.1))])

x0 = np.array([2.8,-2.8])
res = minimize(Rosenbrock, x0,
               method='nelder-mead',options={'xtol': 1e-8, 'disp': True})
print(res.x)

#np.random.seed(46378)
from sklearn.gaussian_process import GaussianProcessRegressor 
from sklearn.gaussian_process.kernels import ConstantKernel, RBF
kernel = ConstantKernel(1.0, (1e-3, 1e3)) * RBF(10, (1e-2, 1e2))

gpr = GaussianProcessRegressor(kernel = kernel,  n_restarts_optimizer = 9)
gpr.fit(x_train,y_train)

y_pred = np.zeros((grid_x.shape[1], grid_x.shape[1]))

for i in range (len(grid_x)):
    for j in range (len(grid_x)):
        y_pred[i][j]=gpr.predict(np.array([grid_x[i][j],grid_y[i][j]]).reshape(1,-1))[0]

from matplotlib import cm
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(grid_x, grid_y, y_pred, cmap=cm.viridis, linewidth=0, antialiased=False, alpha = 0.3)
ax.scatter(x_train[:,0],x_train[:,1],y_train, color = 'red', s=40, marker = '+')
plt.show()

x_itter=[]
y_itter=[]
likelihood = []

y = 1000

dif = 1000

max_iter=50
max_stay=10
dif_min = 10**(-8)
iflow = False
iterlow = 0

#x0=np.array([2,-2,])
fun = lambda x: gpr.predict(x.reshape(1,-1))

for i in range (max_iter):
    y_prev = y
    
    res = minimize(fun, x0,
               method='TNC',bounds=((-3., 3.), (-3., 3.)))
    x_itter.append(x0)
    x0 = res.x    
    y = fun(x0)
    print(y)
    y_itter.append(y)
    dif = abs(y - y_prev)
    
    # Проверка, не застряли ли мы слишком надолго в яме (взято из программы Алексея)
    if dif < dif_min:
        if (iflow==False):
            iterlow=i
            iflow=True
    # Если застряли - то надо выходить
    if ((iflow == True) and (i - iterlow > max_stay)):
        break
    #print('iteration',i,' dif ',dif,x0)
    print('iteration',i, ' dif ',dif,x0,y)
    # Добавляем новую точку в train data
    x_train=np.append(x_train,x0.reshape(1,-1),axis=0)
    y = Rosenbrock(x0)
    y_train=np.append(y_train,np.array(y))
    gpr.fit(x_train, y_train)
    likelihood.append(gpr.log_marginal_likelihood_value_)
```
